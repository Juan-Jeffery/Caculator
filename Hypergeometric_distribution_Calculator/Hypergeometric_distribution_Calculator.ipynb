```python
# @title 超幾何分佈！！ 計算機

import matplotlib.pyplot as plt
from math import comb

def hypergeometric(N, K, n, k):
    return (comb(K, k) * comb(N - K, n - k)) / comb(N, n)

def main():
    #Input section
    N = int(input('總體大小 (N)：'))
    K = int(input('總體中成功的數量 (K)：'))
    n = int(input('抽樣數量 (n)：'))
    k = int(input('抽樣中成功的數量 (k)：'))

    #Calculation section
    result = hypergeometric(N, K, n, k)

    #Output section
    print('超幾何分佈的概率是:', result)

    #Calculate the probability distribution for all possible k values
    k_values = range(max(0, n + K - N), min(K, n) + 1)
    probabilities = [hypergeometric(N, K, n, i) for i in k_values]

if __name__ == '__main__':
    main()
```
