```python
# @title 飛起來！！ 計算機
import math

def calculate():
    ########Input########
    weight = float(input('體重(kg) ＝ '))
    thigh_length = float(input('大腿長度(cm) ＝ '))
    thigh_circumference = float(input('大腿肚圍(cm) ＝ '))
    calf_length = float(input('小腿長度(cm) ＝ '))
    calf_circumference = float(input('小腿肚圍(cm) ＝ '))
    instep_length = float(input('腳背長度(cm) ＝ '))
    instep_circumference = float(input('腳背寬度(cm) ＝ '))
    ########Calculation########
    The_root_sign_of_weight = math.sqrt(weight)
    #大腿正面表面積(m2)
    thigh_front_surface_area_m2 = ((thigh_length * thigh_circumference) + 2 * ((thigh_circumference ** 2) / 4 * math.pi)) * 0.0001 / 2
    #小腿正面表面積(m2)
    Calf_front_surface_area_m2 = ((calf_length * calf_circumference) + 2 * ((calf_circumference ** 2) / 4 * math.pi)) * 0.0001 / 2
    #腳背表面積(m2)
    Instep_surface_area = (instep_length * instep_circumference) * 0.0001
    #兩隻腳打水正面表面積和(m2)
    The_sum_of_the_front_surface_areas_of_two_feet_pumping_water = thigh_front_surface_area_m2 * 2 + Calf_front_surface_area_m2 * 2 + Instep_surface_area * 2
    #
    #空氣密度(g/cm3)
    Air_density = 0.0012
    #水密度(g/cm3)
    water_density = 1.0025
    #空氣密度/水密度後開根號
    water_density_conversion = math.sqrt(Air_density / water_density)
    #
    #通用振翅頻率(Hz)
    Wing_beat_frequency = The_sum_of_the_front_surface_areas_of_two_feet_pumping_water / The_root_sign_of_weight
    #打水頻率(Hz)
    Water_pumping_frequency = (The_sum_of_the_front_surface_areas_of_two_feet_pumping_water / The_root_sign_of_weight) * water_density_conversion

    ########Output########
    print('每秒需', Wing_beat_frequency, '次擺動才能起飛！')
    print('每秒需', Water_pumping_frequency, '次擺動才能跟鯨魚游的一樣快！')

#Run
calculate()
```
